```python
import anthropic
import base64
from groq import Groq
from pathlib import Path
from dotenv import load_dotenv
import requests
import os
import httpx
import sys
from langtrace_python_sdk import langtrace, inject_additional_attributes

load_dotenv(dotenv_path=Path.cwd().parent / ".env")
langtrace.init(api_key = os.getenv("LANGTRACE_API_KEY"),  write_spans_to_console=False) #toggle
claude_key = os.getenv("CLAUDE_API_KEY")
```

```python
parent_dir = Path("..").resolve()
sys.path.append(str(parent_dir))

print("Added to sys.path:", parent_dir)
from helpers.images import image_file_to_base64, image_file_to_data_url, get_latest_image
```

```python
CAPTURE_DIR = os.getenv('CAPTURE_DIRECTORY')
os.makedirs(CAPTURE_DIR, exist_ok=True)

image_path = get_latest_image(f"{CAPTURE_DIR}*")
media_type, image_b64 = image_file_to_base64(image_path)


client = anthropic.Anthropic(api_key=claude_key)

message = client.messages.create(
    model="claude-sonnet-4-5",  # or latest claude-sonnet-4-5 model name you have
    max_tokens=1024,
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image",
                    "source": {
                        "type": "base64",
                        "media_type": media_type,
                        "data": image_b64,
                    },
                },
                {
                    "type": "text",
                    "text": os.getenv("OPEN_AI_PROMPT"),
                },
            ],
        }
    ],
)

print(message)
```
